```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress
```

```python
child_original = pd.read_csv('../../Derivatives/childhood/original_motion_1mm.csv')
adult_original = pd.read_csv('../../Derivatives/adult/original.csv')
hcpd = pd.read_csv('../../Derivatives/childhood/Hurst_motion_schaefer400.csv')
```

```python
from scipy.stats import pearsonr, partial_corr
import matplotlib.pyplot as plt
import pandas as pd

# Loop through genes
for gene in ['PV', 'SST', 'VIP', 'GAD1', 'GAD2']:
    figure = plt.figure()
    
    # Read gene data
    df_gene = pd.read_csv(f'../../Dataset/adult/RNA/schaefer_{gene}.csv', header=None)
    
    # Scatter plot
    plt.scatter(df_gene[0], adult_original.H, c=child_original.SArank, cmap='plasma_r')
    
    # Extract values for analysis
    x = df_gene[0].values
    y = adult_original.H.values
    mediator = adult_original.SArank.values
    
    # Calculate Pearson correlation
    r_value, p_value = pearsonr(x, y)
    
    # Perform partial correlation test
    partial_corr_test = partial_corr(data=pd.DataFrame({'x': x, 'y': y, 'mediator': mediator}), 
                                     x='x', y='y', covar='mediator')
    partial_r = partial_corr_test['r']
    partial_p = partial_corr_test['p-val']
    
    # Display results on the plot
    plt.text(0.05, 0.95, f'R = {r_value:.2f}, P = {p_value:.2e}\n'
                          f'Partial R = {partial_r:.2f}, Partial P = {partial_p:.2e}', 
             transform=plt.gca().transAxes, fontsize=12, verticalalignment='top', 
             bbox=dict(boxstyle="round", facecolor="white"))
    
    # Add labels, legend, and save the plot
    plt.xlabel('Gene Expression')
    plt.ylabel('Child H')
    plt.title(f'{gene} Correlation')
    plt.colorbar(label='SArank')
    plt.savefig(f'CBPD_child_{gene}.eps')
    plt.close()
```

```python
for gene in ['PV', 'SST', 'VIP', 'GAD1', 'GAD2']:
    figure = plt.figure()
    df_gene = pd.read_csv('../../Dataset/adult/RNA/schaefer_'+gene+'.csv', header=None)
    plt.scatter(df_gene[0], child_original.H, c=child_original.SArank, cmap='plasma_r')
    x = df_gene[0].values
    y = child_original.H.values
    slope, intercept, r_value, p_value, std_err = linregress(x, y) 
    n = len(x)
    df_regression = 1
    df_residual = n - 2
    print(df_residual)
    r_squared = r_value**2
    F_value = (r_squared / df_regression) / ((1 - r_squared) / df_residual)
    regression_y = slope * x + intercept
    plt.plot(x, regression_y, color='black', linewidth=2, label='Regression Line')
    # Display R and p values on the plot
    plt.text(0.05, 0.95, f'R = {r_value:.2f}\nF = {F_value:.5e}\nP = {p_value:.2e}', 
             transform=plt.gca().transAxes, fontsize=12, verticalalignment='top', bbox=dict(boxstyle="round", facecolor="white"))
    # Add labels, legend, and save the plot
    plt.savefig(f'CBPD_child_{gene}.eps')
```
