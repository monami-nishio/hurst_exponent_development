```python
import pandas as pd
import numpy as np
import os
import statsmodels.api as sm
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import Pipeline
from natsort import natsorted
from scipy.stats import pearsonr
import seaborn as sns
import matplotlib.cm as cm
from pygam import LinearGAM, s, f
from scipy import stats
sns.set(font_scale=1.5, style="white")
```

```python
indv_adult = pd.read_csv('../Derivatives/adult/H.csv')
motion = pd.read_csv('../Derivatives/adult/plasticity_mri_sample.csv')
```

```python
FD = []
for x in indv_adult['H_all401'].values:
    FD.append(motion[motion.record_id==x.replace('sub-', '').replace('A', '')]['rest_A_motion'].values[0])
```

```python
indv_adult['meanFD'] = FD
indv_adult.to_csv('../Derivatives/adult/hurst_motion.csv')
```

```python
indv = pd.read_csv('../Derivatives/childhood/H.csv')
indv = indv.iloc[:-16, :]
nonormalize = pd.read_csv('../Dataset/childhood/dat/nonormalize.csv')
qa = pd.read_csv('../Dataset/QA_revision_loose.csv')
```

```python
indv['sub'] = indv['H_all401']
indv['run'] = [int(x.replace('run-0', '')) for x in indv['H_all402']]
```

```python
pd.merge(indv, qa, on=['sub', 'run'], how='inner')
```

```python
indv
```

```python
pd.merge(indv, qa, left_on='H_all401', right_on='sub', how='inner')
```

```python
check_motion = pd.merge(indv, qa, left_on='H_all401', right_on='sub', how='inner')
```

```python

check_motion = pd.merge(check_motion, nonormalize, left_on='sub', right_on='record_id', how='inner')
check_motion['Mean'] = check_motion.iloc[:, :400].mean(axis=1)
check_motion['Age_round'] = [round(x) for x in check_motion.age_scan]
```

```python
check_motion = check_motion.loc[:, ['record_id', 'Age_round', 'meanFD', 'Mean']].groupby('record_id').mean().reset_index()
```

```python
for x in np.arange(4,11):
    figure = plt.figure()
    plt.scatter(check_motion[check_motion.Age_round==x].meanFD, check_motion[check_motion.Age_round==x].Mean)
    df_x = pd.DataFrame(check_motion[check_motion.Age_round==x].meanFD.values)
    df_y = pd.DataFrame(check_motion[check_motion.Age_round==x].Mean.values)
    mod = LinearRegression()
    mod_lin = mod.fit(df_x, df_y)
    y_lin_fit = mod_lin.predict(df_x)
    r2_lin = mod.score(df_x, df_y)
    plt.plot(df_x[0].values, [y[0] for y in y_lin_fit], color = '#000000', linewidth=2)
    X = sm.add_constant(check_motion[check_motion.Age_round==x].meanFD.values)
    results = sm.OLS(check_motion[check_motion.Age_round==x].Mean.values, X).fit()
    print(results.summary())
```

```python
plt.scatter(check_motion.meanFD, check_motion['Mean'])
df_x = pd.DataFrame(check_motion.meanFD.values)
df_y = pd.DataFrame(check_motion.Mean.values)
mod = LinearRegression()
mod_lin = mod.fit(df_x, df_y)
y_lin_fit = mod_lin.predict(df_x)
r2_lin = mod.score(df_x, df_y)
plt.plot(df_x[0].values, [y[0] for y in y_lin_fit], color = '#000000', linewidth=2)
X = sm.add_constant(check_motion.meanFD.values)
results = sm.OLS(check_motion.Mean.values, X).fit()
print(results.summary())
plt.savefig('correlation_with_motion.eps')
```
