```python
# exclusion
#Using data from the remaining sample (n=1,052), we identified fluctuation amplitude outliers at the regional level based on a cutoff of 4s.d. from the mean. Individuals with outlier data in more than 5% of cortical regions (n=19) were excluded, producing the final study sample of 1,033 individuals.
exclusion = []
thresholds = []
for i in np.arange(400):
    thr = part.iloc[:, i]<0.1
    std = np.std(part.iloc[:, i])
    mean = np.mean(part.iloc[:, i])
    mins=mean-4*std
    maxs=mean+4*std
    exclusion.append([1 if (x<mins)|(x>maxs) else 0 for x in part.iloc[:,i] ])
    thresholds.append(thr)
outlier = pd.DataFrame(pd.DataFrame(exclusion).sum())#>400*0.05)
outlier[outlier[0]==True]
thresholds = pd.DataFrame(thresholds)
plt.scatter(SAaxis['finalrank.wholebrain'].values, thresholds.sum(axis=1), s=5)
plt.title('parcel with outlier samples')
plt.xlabel('parcels')
figure = plt.figure()
plt.scatter(np.arange(outlier.shape[0]), outlier[0], s=5)
plt.title('sample with outlier parcels')
plt.xlabel('samples')
```
